```python
# Polynomial Regression
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100,1)
y = 2 + 3*x + 5*x**2 + np.random.rand(100,1)

plt.plot(x, y, 'o')

#Model: y=w0 + w1*x + w2*x**2
#Choose w0 = w1 = w2 = 0

w0, w1, w2, w0_temp, w1_temp, w2_temp = 0, 0, 0, 0, 0, 0
converged = False

alpha = 0.1
m = 100
epsilon = 0.000001

while not converged:
    #param = param - (alpha/m)*(derivative summation)
    w0_temp = w0 - np.sum((alpha/m)*(w0 + w1*x + w2*x**2 - y))
    w1_temp = w1 - np.sum((alpha/m)*(w0 + w1*x + w2*x**2 - y)*x)
    w2_temp = w2 - np.sum((alpha/m)*(w0 + w1*x + w2*x**2 - y)*x**2)
    
    if abs(w1_temp - w1) < epsilon and abs(w0_temp - w0) < epsilon and abs(w2_temp - w2):
        converged = True
        
    w0 = w0_temp
    w1 = w1_temp
    w2 = w2_temp

sq = (w0 + w1*x + w2*x**2 - y)**2
cost = np.sum(sq)
print("cost: %f" %cost)
    
print("w0 = %f" %w0)
print("w1 = %f" %w1)
print("w2 = %f" %w2)
plt.plot(x, w0 + w1*x + w2*x**2, 'ro')
plt.show()
```

```python
# Linear Regression using libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.random.rand(100,1)
y = 2 + 3*x + np.random.rand(100,1)

plt.plot(x, y, 'o')

reg = LinearRegression().fit(x, y)
y_pred = reg.predict(x)
plt.plot(x, y_pred)

sq = (y_pred - y)**2
cost = np.sum(sq)
print("cost: %f" %cost)

print("w0 = %f" %reg.intercept_[0])
print("w1 = %f" %reg.coef_[0][0])

plt.show()
```

```python
# Linear Regression using Normal Equation
import numpy as np
import matplotlib.pyplot as plt

x0 = np.ones((100,1))
x1 = np.random.rand(100,1)
x2 = x1**2
y = 10*x0 + 2*x1 + 9*x2 + np.random.rand(100,1)

X = np.concatenate((x0, x1), axis=1)
X = np.concatenate((X, x2), axis=1)
X_t = X.transpose()
P = np.linalg.pinv(np.matmul(X_t, X))
W = np.matmul(np.matmul(P, X_t), y)

y_pred = W[0][0] + W[1][0]*x1 + W[2][0]*x2
plt.plot(x1, y, 'ro')
plt.plot(x1, y_pred, 'o')
plt.show()
```
